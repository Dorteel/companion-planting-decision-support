```python
import pandas as pd
df = pd.read_csv('companion-planting.csv', header=None)

names = [w for s in set(df[0]) | set(df[1]) for w in s.split('/')]

import requests
def get_wdata(name):
    for info in requests.get(
        'https://www.wikidata.org/w/api.php',
        params={
            'action':'wbsearchentities',
            'search':name,
            'language':'en',
            'format':'json'
        }
    ).json()['search']:
        if 'description' in info:
            d = info['description']
            if 'plant' in d or 'species' in d:
                return name, info['id'], d
    return name, None, None

from tqdm.notebook import tqdm
wdatas = [get_wdata(n) for n in tqdm(names)]
```

```python
' '.join(['wd:'+i for _, i, _ in wdatas if i])
```

```python
id_latin = {i.replace('http://www.wikidata.org/entity/',''):l
            for i,l in pd.read_csv('id-latin.csv').values}

pd.DataFrame([
    (name, i or '', id_latin.get(i, ''))
    for name, i, _ in wdatas
], columns=['name', 'id', 'latin']).to_csv('companion-planting-wikidata-latin.csv', index=None)
```
